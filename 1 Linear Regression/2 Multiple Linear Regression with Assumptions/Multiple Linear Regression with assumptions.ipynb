### Dependent and independent variables:
We will use **.iloc()** function we have dataframe not numpy array.

```python
X = dataset.iloc[:,:-1] 
y = dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

random_state is basically a seed for the random number generator.

It ensures that every time you run this code, you get the same train-test split.

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Model Training:

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
regression.coef_     #we have 8 features. so we have 8 intercepts
```

```python
regression.intercept_     #and only one intercept
```

### Prediction for the test data:

```python
y_pred = regression.predict(X_test)
```

```python
y_pred
```

### Performance metics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```

### R2 and Adjusted R2:

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
print(score)
```

```python
#Adjusted R2:
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

### Pickling:-

Pickling in Python means saving a Python object to a file, so you can load it later exactly as it is.

The opposite of pickling is unpickling (reading it back).

It’s mostly used to save models, DataFrames, dictionaries, or any Python object.

#### Why we use it in ML:--

Imagine you train a regression model — it took 10 minutes.
Instead of training it again every time, you pickle it and load it whenever you want.

```python
import pickle
pickle.dump(regression,open('regressor.pkl','wb'))
```

```python
model = pickle.load(open('regressor.pkl','rb'))
```

```python
model.predict(X_test)
```
